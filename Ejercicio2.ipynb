```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos_usuarios = {
    'ID': [101, 102, 103, 104, 105],
    'nombre': ['Andres Reyes', 'Emmanuel Fernandes', 'Diego Gonzales', 'Ana Morillo', 'Josmary Valero'],
    'correo': [
        'Andersreyes10@gmail.com', 
        'ennanuelfernande@provider.net', 
        'Diegoerggonzales22@webmail.org', 
        'Ana.Morrillo@gmail.com', 
        'fandelonepiece04009@company.com'
    ],
    'calificacion': [18, 14, 19, 17, 15],
    'estado': ['activo', 'inactivo', 'activo', 'activo', 'inactivo']
}
```

```python
df = pd.DataFrame(datos_usuarios)
```

```python
usuarios_top = df[df['calificacion'] > 17]
```

```python
plt.figure(figsize=(8, 5))
plt.bar(usuarios_top['nombre'], usuarios_top['calificacion'], color='skyblue', edgecolor='navy')
```

```python
plt.title('usuarios con calificación Superior a 17', fontsize=14)
plt.xlabel('nombre del usuario', fontsize=12)
plt.ylabel('puntuación de calificación', fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
